```python
import altair as alt
import pandas as pd
```

```python
URL = 'https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv'
df = pd.read_csv(URL)
df
```

```python
brush = alt.selection_interval(encodings=['x','y'])
chart1 = alt.Chart.from_dict({
    "data": {"url": "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"},
    "height": 300,
    "width": 500,
    "mark": "bar",
    "encoding": {
      "x": {
        "field": "Total Floors",
        "type": "quantitative"
      },
      "y": {
        "aggregate": "mean",
        "field": "Square Footage",
        "type": "quantitative"
      }
    }
}).add_selection(
    brush
)
chart2 = alt.Chart.from_dict({
    "data": {"url": "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"},
    "height": 300,
    "width": 500,
    "mark": "rule",
    "encoding": {
      "y": {
        "aggregate": "mean",
        "field": "Square Footage",
        "type": "quantitative"
      },
      "color": {"value": "firebrick"},
      "size": {"value": 3}
    }
}).transform_filter(
    brush
)
chart = alt.layer(chart1, chart2)
chart
```

```python
chart3 = alt.Chart.from_dict({ 
    "width": 600, 
    "height": 500,
    "data": {"url": "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"},
    "mark": "area",
    "params": [{
        "name": "Floors",
        "select": {"type": "point", "fields": ["Total Floors"]},
        "bind": "legend"
    }],
    "encoding": {
        "x": {
          "timeUnit": "year", "field": "Year Constructed",
          "axis": {"domain": False, "format": "%Y", "tickSize": 0}
        },
        "y": {
          "aggregate": "sum", "field": "Square Footage",
          "stack": "center", "axis": {"title": "Total Square Footage"}
        },
        "color": {
              "field":"Total Floors",
              "scale": {"scheme": "category20b"}
        },
        "opacity": {
              "condition": {"param": "Floors", "value": 1},
              "value": 0.2
        }
    }
})
chart3
```
